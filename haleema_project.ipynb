```python
# seaborn histogram 
sns.catplot(x="BuildingType", kind="count",height=6, aspect=2,data=df1)
#Catplot Python Seaborn: One Function to Rule All Plots With Categorical Variables
```

```python
#seaborn Scatterplot
#(z-score )normalization is applied on "TotalGHGEmissions"
plt.figure(figsize=(15,7))
df1["GHGEmissions(MetricTonsCO2e)"]=(df1["GHGEmissions(MetricTonsCO2e)"]-df1["GHGEmissions(MetricTonsCO2e)"].mean())/df1["GHGEmissions(MetricTonsCO2e)"].std()
sns.scatterplot(x=df1['BuildingType'], y=df1['GHGEmissions(MetricTonsCO2e)'])
#NonResidentials buildings are higher in GHGEmissions(MetricTonsCO2e)
```

```python
df1["GHGEmissionsIntensity(kgCO2e/ft2)"]=(df1["GHGEmissionsIntensity(kgCO2e/ft2)"])/df1["GHGEmissionsIntensity(kgCO2e/ft2)"].max()
plt.figure(figsize=(15,8))
sns.lineplot(x='BuildingType',y='GHGEmissionsIntensity(kgCO2e/ft2)', data=df1)
#line_plot
```

```python
plt.figure(figsize=(15,10))
df1["GHGEmissions(MetricTonsCO2e)"]=(df1["GHGEmissions(MetricTonsCO2e)"])/df1["GHGEmissions(MetricTonsCO2e)"].max()
sns.boxplot(y ='GHGEmissions(MetricTonsCO2e)',data= df1, x= 'BuildingType')
plt.xlabel('BuildingType')
## Multifamily MR (5-9),SPS-District K-12 , Multifamily LR (1-4), have almost same mean
```

```python
plt.scatter(df1["NaturalGas(kBtu)"],df1["GHGEmissions(MetricTonsCO2e)"],  color='blue')
plt.xlabel("NaturalGas(kBtu)")
plt.ylabel("GHGEmissions(MetricTonsCO2e)")
plt.show()
```
